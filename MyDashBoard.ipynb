```python
def fig(df , element , to_ = None ):
    if not to_ : 
        to_  = (df[element]>0).sum()
    df2 = df.where(df[element]>0).dropna().sort_values(element, ascending = False)[:to_]
    df2['{} per kcal'.format(element)] =(df2[element]-df2[element].min())/(df2[element].max() - df2[element].min())
    fig = px.scatter(df2,y=element ,x='Energy (kcal)',color='Food Group',
               hover_name='ShortDescrip', size = '{} per kcal'.format(element),
               width=1000, height=500,log_x=True, template='seaborn')
    return fig
```

```python
app.layout = html.Div(
    [
        title,
        dbc.Form(html.Div(children=[card1,card2,card3,card4,card5])),
        element_fig,
        pic,
        BMI,
        element_fig3,
        element_fig2,
        end
    ])


@app.callback(
    Output('unit','options'),
    Input('element', 'value'),
    
)

def update_Div( element):
    if element in only_SI:
        u = ['SI']
    else:
        u = ['SI', 'USRDA']
    return u

@app.callback(
    Output('contain_element', 'figure'),
    Input('element', 'value'),
    Input('n_item' , 'value'),
    Input('unit','value'),
    Input('keto','n_clicks'),
    Input('low','n_clicks'),
    Input('car','n_clicks'),
    Input('veg1','n_clicks'),
    Input('veg2','n_clicks'),
    Input('all','n_clicks'),
)
def draw_fig (element, n_item, unit,ket, lw ,cr,veg1,veg2, all1):
    df = df_source.copy()
    if ket and ket > clicks[0]:
        clicks[0] = ket 
        df = df_source.where(df_source['group'].apply(lambda x :x in keto)).dropna()
    elif lw and lw > clicks[1]:
        clicks[1] = lw 
        df = df_source.where(df_source['group'].apply(lambda x :x in low)).dropna()
    elif cr and cr > clicks[2]:
        clicks[2] = cr
        df = df_source.where(df_source['group'].apply(lambda x :x in car)).dropna()
    elif veg1 and  veg1 > clicks[3]:
        clicks[3] = veg1 
        df = df_source.where(df_source['group'].apply(lambda x :x in vegan)).dropna()
    elif veg2 and veg2 > clicks[4]:
        clicks[4] = veg2 
        df = df_source.where(df_source['group'].apply(lambda x :x in vegeterian)).dropna()
    elif all1 and all1 > clicks[5]:
        clicks[5] = all1
        df = df_source.copy()
    figure = fig(df, elements[element][unit], n_item)
    return figure    

@app.callback(
    Output('img', 'src'),
    Output('txt', 'children'),
    Output('title', 'children'),
    Input('element', 'value'),
)

def ch_img(element):
    return app.get_asset_url('{}.jpg'.format(element)), 'Food items with high {}'.format(element), element

@app.callback(
    Output('txt2','children'),
    Output('vit_min_plt','figure'),
    Input('vit_min','value'),
)
def vit_plt(e):
    return '{} per Food Groups'.format(e),  avg(fd4,elements[e]['USRDA'])

@app.callback(
    Output('txt1','children'),
    Output('plan_pie','figure'),
    Output('article','children'),
    Input('plan','value'),
)
def pieplt(e):
    return 'Plan Suggestion for: {}'.format(e), sugg(sug,e), art[e]

```
